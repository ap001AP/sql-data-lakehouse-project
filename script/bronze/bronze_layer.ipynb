# Define Ingestion Configuration

```python

files_to_ingest = {
    "/Volumes/workspace/bronze/source_systems/source_crm/sales_details.csv": "workspace.bronze.crm_sales_details",
    "/Volumes/workspace/bronze/source_systems/source_crm/cust_info.csv": "workspace.bronze.crm_cust_info",
    "/Volumes/workspace/bronze/source_systems/source_crm/prd_info.csv": "workspace.bronze.crm_prd_info",
    "/Volumes/workspace/bronze/source_systems/source_erp/CUST_AZ12.csv": "workspace.bronze.erp_CUST_AZ12",
    "/Volumes/workspace/bronze/source_systems/source_erp/LOC_A101.csv": "workspace.bronze.erp_LOC_A101",
    "/Volumes/workspace/bronze/source_systems/source_erp/PX_CAT_G1V2.csv": "workspace.bronze.erp_PX_CAT_G1V2",
}
```

# Ingest Files into Bronze Tables

```python
for file_path, table_name in files_to_ingest.items():
    try:
        df = spark.read.csv(file_path, header=True, inferSchema=True)
        df.write.mode("overwrite").format("delta").saveAsTable(table_name)
        print(f"Loaded {file_path} into {table_name}")
    except Exception as e:
        print(f"Failed on {table_name}: {e}")
```
